```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
sales=pd.read_csv('Amazon Sales data.csv')
sales.head()
```

```python
sales.shape
```

```python
sales.describe()
```

```python
sales.info()
```

```python
sales.isnull().sum()
```

```python
# Convert 'Order Date' and 'Ship Date' to datetime
sales['Order Date'] = pd.to_datetime(sales['Order Date'], errors='coerce')
sales['Ship Date'] = pd.to_datetime(sales['Ship Date'], errors='coerce')
```

```python
sales['Order Date']
```

```python
sales['Ship Date']
```

```python
sales['Year'] = sales['Order Date'].dt.year
sales['Month'] = sales['Order Date'].dt.month
```

```python
sales.head()
```

```python
# Aggregate total revenue month-wise
monthly_sales = sales.groupby(['Year', 'Month'])['Total Revenue'].sum().reset_index()
```

```python
monthly_sales
```

```python
# Aggregate total revenue year-wise
yearly_sales = sales.groupby('Year')['Total Revenue'].sum().reset_index()
```

```python
yearly_sales
```

```python
yearly_monthly_sales = sales.pivot_table(values='Total Revenue', index='Year', columns='Month', aggfunc='sum').fillna(0)
```

```python
yearly_monthly_sales
```

```python
# Plotting the sales trends
plt.figure(figsize=(14, 7))

# Plot monthly sales

monthly_sales_pivot = monthly_sales.pivot(index='Year', columns='Month', values='Total Revenue')
monthly_sales_pivot.plot(ax=plt.gca(), marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.legend(title='Month', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Plot yearly sales
plt.figure(figsize=(14, 7))
plt.plot(yearly_sales['Year'], yearly_sales['Total Revenue'], marker='o')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()
```

```python
# Plot yearly month-wise sales
plt.figure(figsize=(14, 7))
yearly_monthly_sales.T.plot(ax=plt.gca(), marker='o')
plt.title('Yearly Month-wise Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Sales Channel Analysis
sales_channel_revenue = sales.groupby('Sales Channel')['Total Revenue'].sum()
sales_channel_units_sold = sales.groupby('Sales Channel')['Units Sold'].sum()

# Bar Chart for Total Sales by Sales Channel
plt.figure(figsize=(10, 6))
sns.barplot(x=sales_channel_revenue.index, y=sales_channel_revenue.values, palette="pastel")
plt.title('Total Sales by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()
```

```python
# Bar Chart for Total Units Sold by Sales Channel
plt.figure(figsize=(10, 6))
sns.barplot(x=sales_channel_units_sold.index, y=sales_channel_units_sold.values, palette="pastel")
plt.title('Total Units Sold by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Total Units Sold')
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot of Total Profit vs. Units Sold
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sales, x='Units Sold', y='Total Profit', hue='Item Type', palette="deep", s=100)
plt.title('Total Profit vs. Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Total Profit')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Line Chart of Total Profit Over Time (Monthly)
sales['Order Date'] = pd.to_datetime(sales['Order Date'])
monthly_profit = sales.groupby(sales['Order Date'].dt.to_period('M'))['Total Profit'].sum()

plt.figure(figsize=(12, 6))
monthly_profit.plot(kind='line', marker='o')
plt.title('Total Profit Over Time')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Bar Chart of Average Total Revenue by Order Priority
avg_revenue_priority = sales.groupby('Order Priority')['Total Revenue'].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_revenue_priority.index, y=avg_revenue_priority.values, palette="coolwarm")
plt.title('Average Total Revenue by Order Priority')
plt.xlabel('Order Priority')
plt.ylabel('Average Total Revenue')
plt.tight_layout()
plt.show()
```

```python
# Bar Chart of Average Total Profit by Order Priority
avg_profit_priority = sales.groupby('Order Priority')['Total Profit'].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_profit_priority.index, y=avg_profit_priority.values, palette="coolwarm")
plt.title('Average Total Profit by Order Priority')
plt.xlabel('Order Priority')
plt.ylabel('Average Total Profit')
plt.tight_layout()
plt.show()
```

```python
# Top 10 Countries by Sales Revenue
top_10_countries_revenue = sales.groupby('Country')['Total Revenue'].sum().sort_values(ascending=False).head(10)

# Bar Chart for Top 10 Countries by Sales Revenue
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_countries_revenue.index, y=top_10_countries_revenue.values, palette="viridis")
plt.title('Top 10 Countries by Sales Revenue')
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top 10 Countries by Sales Revenue (Online)
top_10_countries_online = sales[sales['Sales Channel'] == 'Online'].groupby('Country')['Total Revenue'].sum().sort_values(ascending=False).head(10)

# Top 10 Countries by Sales Revenue (Offline)
top_10_countries_offline = sales[sales['Sales Channel'] == 'Offline'].groupby('Country')['Total Revenue'].sum().sort_values(ascending=False).head(10)

# Bar Chart for Top 10 Countries by Sales Revenue (Online)
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_countries_online.index, y=top_10_countries_online.values, palette="Blues_d")
plt.title('Top 10 Countries by Online Sales Revenue')
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bar Chart for Top 10 Countries by Sales Revenue (Offline)
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_countries_offline.index, y=top_10_countries_offline.values, palette="Reds_d")
plt.title('Top 10 Countries by Offline Sales Revenue')
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Total Revenue
total_revenue = sales['Total Revenue'].sum()

# Average Revenue per Order
average_revenue_per_order = sales['Total Revenue'].mean()

# Monthly Revenue
monthly_revenue = sales.groupby(sales['Order Date'].dt.to_period('M'))['Total Revenue'].sum()

# Average Order Value
average_order_value =sales.groupby('Order ID')['Total Revenue'].sum().mean()

print(f'Total Revenue: ${total_revenue:.2f}')
print(f'Average Revenue per Order: ${average_revenue_per_order:.2f}')
print('Monthly Revenue:')
print(monthly_revenue)
print(f'Average Order Value: ${average_order_value:.2f}')
```
